```python
# 1. Create a list of names and print the second item.

# Create the list
team_names = ["Cedrick", "Christian Vincent", "Bob Troy", "Bob Tyron"]

# Print the second item (which is at  index 1)
print(f"1. Second item in the list: {team_names[1]}")
```

```python
# 2. Create a list of sports and then replace the second item with another sport.

# Create the list
sports =["Basketball", "Table Tennis", "Sepak Takraw", "Hockey"]

# Replace the second item (index 1)
sports[1] = "Tennis"

print(f"2. List after replacement: {sports}")
```

```python
# 3. Create a list containing numbers and delete the fifth number from the array.

# Create the list
security_codes = [101, 200, 302, 403, 505, 665, 700]

# Delete the fifth number (which is at index 4) using the 'del' keyword
del security_codes[4]

print(f"3. List after deleting the fifth number: {security_codes}")
```

```python
# 4. Create two lists of numbers and merge them.

# Create the two lists
list_a = [10, 20, 30]
list_b = [40, 50, 60]

# Merge the lists using the '+' operator
merged_list = list_a + list_b

print(f"4. Merged list: {merged_list}")
```

```python
# 5. Create a list of numbers and find the length, minimum, and maximum.

# Create the list
data_points = [20, 5, 2, 100, 3]

# Find length, minimum, and maximum using built-in functions
list_length = len(data_points)
list_min = min(data_points)
list_max = max(data_points)

print(f"5. Length: {list_length}, Minimum: {list_min}, Maximum: {list_max}")
```

```python
# 6. Create a dictionary of students and scores and print out a student’s score.

# Create the dictionary
student_scores = {"Carmela": 92, "Bella": 90, "Alexie": 93}

# Print a student's score by referencing their key
carmela_score = student_scores["Carmela"]

print(f"6. Carmela's score: {carmela_score}")
```

```python
# 7. Create a dictionary with the key being names and values being ages and then delete the second key/value pair.

# Create the dictionary
user_ages = {"Cedrick": 14, "Christain Vincent": 21, "Bob Troy": 11, "Bob Tyron": 7}

# Delete the key/value pair using 'del' and the key
del user_ages["Christain Vincent"]

print(f"7.Dictionary after deleting 'Christian Vincent': {user_ages}")
```

```python
# 8. Create a dictionary of names and ages and then print out all the keys and values

# Reusing the dictionary from exercise 7
user_data = {"Cedrick": 14, "Christain Vincent": 21, "Bob Troy": 11, "Bob Tyron": 7}

print("8. Printing all keys and values")
# Print only the keys
print(f"   Keys: {user_data.keys()}")
# Print only the values
print(f"   Values: {user_data.values()}")
```

```python
# 9. Create a tuple of your favorite movies

# Create the tuple
fav_movie = ("The Matrix", "The Amateur", "Naked Gun", "Split")

print(f"9. My favorite movies (Tuple): {fav_movie}")
```

```python
# 10. Create a tuple and print all the items from the first to third index.

# Create the tuple
data_tuple = ("TCP", "UDP", "ICMP", "DNS", "HTTP", "FTP")

# Print items from the first index (index 1) up to, but NOT including, the third index (index 3)
# [start:stop] -> prints index 1 and index 2
slice_result = data_tuple[1:3]

print(f"10. Tuple slice (Index 1 to 3): {slice_result}")
```
